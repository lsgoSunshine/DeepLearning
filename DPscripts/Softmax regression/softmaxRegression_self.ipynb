```python
import torch
from torch import nn
import numpy as np
import sys
sys.path.append('..')
import dlzh.d2lzh_pytorch as d2l
from dlzh import timer
from collections import OrderedDict
from torch.nn import init
```

```python
'''获取和读取数据'''
batch_size = 256
#指定线程数
num_workers = 4
train_iter,test_iter=d2l.load_data_fashion_mnist(batch_size,num_workers)
'''定义和初始化模型'''
#28*28=784,总共10个标签类别
num_inputs=784
num_outputs=10

class softmax(nn.Module):
    #这里的num_inputs和num_outputs其实就是W,b的维度
    def __init__(self,num_inputs,num_outputs):
        super(softmax,self).__init__()
        self.linear=nn.Linear(num_inputs,num_outputs)
    def forward(self,x):
        y=self.linear(x.view(x.shape(0),-1))
        return y
# net= softmax(num_inputs,num_outputs)

net=nn.Sequential()
net.add_module('flatten',d2l.FlattenLayer())
net.add_module('linear',nn.Linear(num_inputs,num_outputs))
print(net[1])
```

```python
'''初始化参数'''
init.normal_(net.linear.weight,mean=0,std=0.01)
init.constant_(net.linear.bias, val=0)
```

```python
'''构造cross entropy算法'''
loss=nn.CrossEntropyLoss()
```

```python
'''定义优化算法'''
Optimizer=torch.optim.SGD(net.parameters(),lr=0.1)

'''训练模型'''
num_epochs=5
d2l.train_ch3(net,train_iter,test_iter,loss,num_epochs,batch_size,None,None,Optimizer)
```
